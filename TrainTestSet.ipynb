## Train-Test method is way to measure the accuracy of a model
## The data set is split into two sets a training set and a testing set.

### 80% of the dataset is split for training set and 20% for testing set.

### Model is trained using the training set and tested using the testing set.



### Let's start with a data set, for example a dataset which illustrates 100 customers in a shop and their shopping habits.

```python
## Example

import matplotlib.pyplot as plt
numpy.random.seed(2)

x = numpy.random.normal(3, 1, 100)
y = numpy.random.normal(150, 40, 100) / x

plt.scatter(x, y)
plt.show()
```

#### Result:
 #### The x axis represents the number of minutes before making a purchase.

 #### The y axis represents the amount of money spent on the purchase.

##### will continue splitting the dataset into training and testing sets.

##### Training set will be 80% of data
##### Testing set will be 20% of remaining data

```python
train_x = x[:80]
train_y = y[:80]

test_x = x[80:]
test_y = y[80:]
```

```python
## Plot the training sets

plt.scatter(train_x, train_y)
plt.show()
```

##### Result
#####  The training set looks similar to the original data set, so it looks like a valid selection of data.

```python
## Lets plot the testing sets also

plt.scatter(test_x,test_y)
plt.show()
```

##### Result
#####  The testing set also looks pretty similar to the original data set with much less values.

##### Fitting the data set

#####   The best fit the data set would be a polynomial regression, lets draw a line of polynomial regression, we make use of the plot() method of the matplotlib.


```python
mymodel = numpy.poly1d(numpy.polyfit(train_x, train_y, 4))

myline = numpy.linspace(0, 6, 100)

plt.scatter(train_x, train_y)
plt.plot(myline, mymodel(myline))
plt.show()
```

##### Result fits the suggestion of the polynomial regression but gives undesired values when we plot points outside the dataset
##### Example: the line indicates that a customer spending 6 minutes in the shop would make a purchase worth 200. That is probably a sign of overfitting.
##### We can make use of the R-squared score to determine how well the data set is fitting the model.

##### The R2 measures the relationship between the x axis and the y axis, and the value ranges from 0 to 1, where 0 means no relationship, and 1 means totally related.

##### The sklearn module has a method called r2_score() that will help us find this relationship.

```python
## We will test how well it fits our model
mymodel = numpy.poly1d(numpy.polyfit(train_x, train_y, 4))

r2 = r2_score(train_y, mymodel(train_x))

print(r2)
```

##### Lets bring in the testing set and do the same exercise for testing set.


```python
from sklearn.metrics import r2_score

mymodel = numpy.poly1d(numpy.polyfit(train_x, train_y, 4))

r2 = r2_score(test_y, mymodel(test_x))

print(r2)
```

##### Check the values for both training sets and testing sets and verify if they are equal, if they are equal then the model fits well.
